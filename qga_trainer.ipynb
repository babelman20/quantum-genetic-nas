### Imports

```python
import numpy as np
import matplotlib.pyplot as plt

from keras.src.datasets import cifar10

from quantum_genetic import QuantumPopulationManager
```

### Get Data

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)

print(X_train[0])
```

### Make Genetic Manager

```python
num_generations = 15

genetic_manager = QuantumPopulationManager(X_train.shape[1:], num_generations)
```

### Run Evolution

```python
data = (X_train, y_train, X_test, y_test)

best_architecture, best_fitness, generational_scores = genetic_manager.run_evolution(data)
```

### Print results

```python
print(best_architecture)
print(best_fitness)
```

```python
generations = range(1, num_generations+1)

top_fitness = [t[0] for t in generational_scores]
mean_fitness = [t[1] for t in generational_scores]
weighted_top_fitness = [t[2] for t in generational_scores]
top_acc = [t[3] for t in generational_scores]
mean_acc = [t[4] for t in generational_scores]
weighted_top_acc = [t[5] for t in generational_scores]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(generations, top_fitness, label='Top Fitness', marker='o')
ax1.plot(generations, mean_fitness, label='Mean Fitness', marker='x')
ax1.plot(generations, weighted_top_fitness, label='Weighted Top Fitness', marker='D')

ax1.set_title('Fitness over Generations')
ax1.set_xlabel('Generation')
ax1.set_ylabel('Fitness')
ax1.legend()
ax1.grid(True)

ax2.plot(generations, top_acc, label='Top Accuracy', marker='o')
ax2.plot(generations, mean_acc, label='Mean Accuracy', marker='x')
ax2.plot(generations, weighted_top_acc, label='Weighted Top Accuracy', marker='D')

ax2.set_title('Accuracy over Generations')
ax2.set_xlabel('Generation')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True)

plt.tight_layout()

plt.show()
```

### Train best model architecture

```python
best_model = genetic_manager.build_model_from_architecture(best_architecture)

best_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = best_model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_test, y_test))
```

### Graph model results

```python
steps = range(1,51)

loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(steps, loss, label='Training Loss')
ax1.plot(steps, val_loss, label='Test Loss')

ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True)

ax2.plot(steps, acc, label='Training Accuracy')
ax2.plot(steps, val_acc, label='Test Accuracy')

ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

### Print out best model archtecture

```python
best_model.summary()
```

### Make confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred_prob = best_model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

y_true = y_test.flatten()
print(y_true.shape, y_pred.shape)

cm = confusion_matrix(y_true, y_pred)

cm_plot = ConfusionMatrixDisplay(confusion_matrix=cm)
cm_plot.plot(xticks_rotation=45, cmap='Blues')
plt.title('CIFAR-10 Confusion Matrix')
plt.show()
```
